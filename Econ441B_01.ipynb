```python
data=pd.read_csv('/content/gdrive/MyDrive/Econ441B/College.csv')
```

```python
data.head()
```

```python
data.insert(loc=1, column="AccRate", value=data['Accept']/data['Apps']*100)
data.head()
```

Apps means application numbers

Accept means how many students have been admitted to the college

I want to see the acceptance rate of all the college and understand the overall situation.

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.title("Histogram of Accept Rate")
sns.histplot(data.AccRate, stat = "density")
sns.kdeplot(data.AccRate, color = "red")

plt.grid()
```

I visualize the histogram of AccRate, which can show the distribution of the overall acceptance rate of the data. Obviously, most of the college have over 60% acceptance rate. And the plot show a left skewness feature.
